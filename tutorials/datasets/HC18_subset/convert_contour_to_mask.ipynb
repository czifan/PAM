```python
import os 
import cv2
import numpy as np
from glob import glob 

def fill_contours(contour_path, output_size=None):
    """
    填充闭合轮廓内部区域
    """
    # 读取图像
    img = cv2.imread(contour_path)
    gray = cv2.cvtColor(img, cv2.COLOR_BGR2GRAY)
    
    # 二值化
    _, binary = cv2.threshold(gray, 127, 255, cv2.THRESH_BINARY)
    
    # 查找轮廓
    contours, _ = cv2.findContours(binary, cv2.RETR_EXTERNAL, cv2.CHAIN_APPROX_SIMPLE)
    
    # 创建空白mask
    if output_size:
        mask = np.zeros(output_size, dtype=np.uint8)
    else:
        mask = np.zeros_like(gray)
    
    # 填充轮廓
    cv2.fillPoly(mask, contours, color=255)
    
    return mask

files = glob("guiding_original_annotations/*.png")
for file in files:
    mask = fill_contours(file)
    cv2.imwrite(os.path.join("guiding_prompts", os.path.basename(file).replace("_Annotation.png", ".png")), mask)
```
